# 1.

Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e=\sum ^{\infty} _{k=0}\frac{1}{k!}\notin \mathbb{Q}$. A factorial is defined as $k!=(k)(k-1)(k-2)(k-3)...,\forall k\in \mathbb{Z}^+$, note that $0!=1$. Furthermore, a sum notation $\sum ^\infty _{k=0} k=0+1+2+3+...+...$

### Proof:

Assume that $e$ is rational.

There exist coprime integers $m$ and $n$ (let $n\in \mathbb{Z}^+$) such that:
$$
\frac{m}{n} = e = \sum ^\infty _{i=0} \frac{1}{i!}
$$

Multiplying both sides by $n!$, note that:
$$
\frac{m}{n}n!=n!\sum ^\infty _{i=0} \frac{1}{i!}=\left(\frac{n!}{0!}+\frac{n!}{1!}+\frac{n!}{2!}+\cdots+\frac{n!}{n!}\right)+\left(\frac{n!}{(n+1)!}+\frac{n!}{(n+2)!}+\cdots\right)
$$

Hence:
$$
m(n-1)!-\left(\frac{n!}{0!}+\frac{n!}{1!}+\frac{n!}{2!}+\cdots+\frac{n!}{n!}\right)=\frac{1}{(n+1)}+\frac{1}{(n+1)(n+2)}+\frac{1}{(n+1)(n+2)(n+3)}+\cdots
$$
$$
\hspace{6cm}<\frac{1}{(n+1)}+\frac{1}{(n+1)(n+1)}+\frac{1}{(n+1)(n+1)(n+1)}+\cdots
$$
$$
=\sum ^\infty _{i=0}\left(\frac{1}{n+1}\right)^{(i+1)}
$$
$$
=\frac{\frac{1}{n+1}}{1-\frac{1}{n+1}}=\frac{1}{n}\leq 1
$$

Note that the quantity on the left hand side is strictly positive, since it is equal to:
$$
\frac{1}{(n+1)}+\frac{1}{(n+1)(n+2)}+\frac{1}{(n+1)(n+2)(n+3)}+\cdots
$$
...which is strictly positive.

Thus,
$$
m(n-1)!-\left(\frac{n!}{0!}+\frac{n!}{1!}+\frac{n!}{2!}+\cdots+\frac{n!}{n!}\right)
$$
must be a strictly positive integer that is less than 1.

Because this is a contradiction, $e$ must be irrational.

# 2.

Prove Minkowski's Inequality for sums $\forall p>1, (a_k, b_k)>0:$
$$\begin{bmatrix}\sum ^n _{k=1}|a_k+b_k|^p\end{bmatrix}^\frac{1}{p}\leq \begin{bmatrix}\sum^n _{k=1}|a_k|^p\end{bmatrix}^\frac{1}{p}+\begin{bmatrix}\sum^n _{k=1}|b_k|^p\end{bmatrix}^\frac{1}{p}$$

### Proof

#### Step 1: Laws of Exponents (Power raised to a power):

- Let $n = $ any base number, then ${(n^p)}^\frac{1}{p}=n$
- We know that ${(n^p)}^\frac{1}{p}=n$ for any base number $n$, which implies that the $p$-th root of $n^p$ is $n$. Applying this to the sums of $|a_k|^p$ and $|b_k|^p$, we have:

\begin{gather} 
\Sigma_{k=1}^n |a_k + b_k|
\leq
\Sigma_{k=1}^n |a_k|
+
\Sigma_{k=1}^n |b_k|
\end{gather}

#### Step 2: Property of Summation (Addition)

\begin{gather} 
\Sigma_{k=1}^n |a_k + b_k|
\leq
\Sigma_{k=1}^n |a_k| + |b_k|
\end{gather}

#### Step 3: Prove by Triangle Inequality

- Disregarding the Sigma Notation we get:
\begin{gather}
|a_k + b_k|
\leq
|a_k| + b_k| 
\end{gather}

- Since, $(a_k, b_k) > 0$, then $(a + b) = a + b$, and $|a| = a_k$, $|b| = b$:
- Substitute:
\begin{gather}
|a + b| = a + b = |a| + |b|
\end{gather}


Thus,

$$\begin{bmatrix}\sum ^n _{k=1}|a_k+b_k|^p\end{bmatrix}^\frac{1}{p}\leq \begin{bmatrix}\sum^n _{k=1}|a_k|^p\end{bmatrix}^\frac{1}{p}+\begin{bmatrix}\sum^n _{k=1}|b_k|^p\end{bmatrix}^\frac{1}{p}$$

# 3.

Prove the triangle inequality $|x+y|\leq |x|+|y|, \forall(x,y)\in\mathbb{R}$

### Proof:
We must first prove that $\left(|x|+|y|\right)^2\geq|x+y|^2$. Note that $\forall n\in\mathbb{R}$, $|n|^2 = n^2$.

$$
\left(|x| + |y|\right)^2 = |x|^2 + 2|x||y| + |y|^2
$$

Because $|x|\geq x$ and $|y|\geq y$:

$$
|x|^2 + 2|x||y| + |y|^2\geq x^2 + 2xy + y^2 = \left(a + b\right)^2 = |a+b|^2
$$

So,
$$
(|a|+|b|)^2\geq |a+b|^2
$$

Therefore, getting the square root of the inequality above,
$$
|a|+|b|\geq|a+b|
$$

That is,
$$
|a+b|\leq|a|+|b|
$$

# 4.

Prove Sedrakayan's Lemma $\forall u_i, v_i\in\mathbb{R}^+$:
$$\frac{(\sum ^n _{i=1}u_i)^2}{\sum ^n _{i=1}v_i}\leq\sum ^n _{i=1}\frac{(u_i)^2}{v_i}$$

### Proof:

#### Step 1:

To prove Sedrakayanâ€™s Lemma directly, we need to demonstrate that for all $(u_i, v_i \in \mathbb{R}^+)$, the inequality

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2 \leq \frac{\sum_{i=1}^n (u_i)^2}{v_i}$$

holds true. 

#### Step 2:

The ***Cauchy-Schwarz Inequality*** states that for any real numbers $(a_i)$ and $(b_i)$, 

$$\left(\sum_{i=1}^n a_i b_i\right)^2 \leq \left(\sum_{i=1}^n (a_i)^2\right) \left(\sum_{i=1}^n (b_i)^2\right)$$

#### Step 3:

Let $a_i = \sqrt{u_i}$ and $b_i = \frac{1}{\sqrt{v_i}}$. Then, the Cauchy-Schwarz Inequality becomes

$$\left(\sum_{i=1}^n \sqrt{u_i} \cdot \frac{1}{\sqrt{v_i}}\right)^2 \leq \left(\sum_{i=1}^n u_i\right) \left(\sum_{i=1}^n \frac{1}{v_i}\right)$$

#### Step 4:

The *left-hand side* simplifies to 

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2$$

#### Step 5:

The *right-hand side* simplifies to 

$$\frac{\sum_{i=1}^n (u_i)^2}{v_i}$$

So,

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2 \leq \frac{\sum_{i=1}^n (u_i)^2}{v_i}$$

That is,

$$\frac{(\sum ^n _{i=1}u_i)^2}{\sum ^n _{i=1}v_i}\leq\sum ^n _{i=1}\frac{(u_i)^2}{v_i}$$

\begin{gather}
\sqrt{|f(x)|^2} \geq \sqrt{h(x)} - \sqrt{|g(x)|^2}, \forall (x,y) \in \mathbb{Z}^+
\end{gather}

With the given functions, we can substitute $f(x)$ and $g(x)$ into $j(x)$:
$$
    j(x) = |(2x + 1) + (2y - 1)| = |2x + 2y| = 2|x + y|.
$$

With this, we can also find $h(x)$:
$$
    h(x) = j(x)^2 = (2|x + y|)^2 = 4(x + y)^2.
$$

We can then simplify the inequality by substituting the function names:
$$
    \sqrt{h(x)} = \sqrt{4(x + y)^2} = 2|x + y|.
$$

$$
    \sqrt{|f(x)|^2} = \sqrt{(2x + 1)^2} = |2x + 1|
$$

$$
    \sqrt{|g(x)|^2} = \sqrt{(2y - 1)^2} = |2y - 1|
$$

So we have:

$$
    |2x + 1| \geq 2|x + y| - |2y - 1|
$$

When analyzing the different absolute values in the equation as well as from the given, we can see that:
- For $( 2x + 1 )$, it's always positive for $( x \in \mathbb{Z}^+ )$, hence $( |2x + 1| = 2x + 1 )$.
- For $( 2y - 1 )$, if $( y \geq 1 )$, then $( |2y - 1| = 2y - 1)$

Therefore, the equality can be simplified down to:
$$
    2x + 1 \geq 2(x + y) - (2y - 1)

$$
    2x + 1 \geq 2x + 2y - 2y + 1

$$
    2x + 1 \geq 2x + 1

2. Prove Sedrakayan's Lemma if $u_i, v_i$ are square roots of an even integers:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

To prove Sedrakayan's Lemma when $ u_i $ and $ v_i $ are square roots of even integers:

Let $ u_i = \sqrt{2a_i} $ and $ v_i = \sqrt{2b_i} $, where $ a_i $ and $ b_i $ are positive integers. We need to show:

$$
\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{(2a_i)}{(2b_i)}
$$

#### Step 1: Expand the terms

Expanding the terms, we get:

$$
\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} = \frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right) \left( \sum_{i=1}^n \sqrt{2a_i} \right)}{\sum_{i=1}^n \sqrt{2b_i}}
, \quad
\sum_{i=1}^n \frac{(2a_i)}{(2b_i)} = \sum_{i=1}^n \frac{2a_i}{2b_i}
$$


#### Step 2: Rewrite

Since $ \sqrt{2a_i} $ and $ \sqrt{2b_i} $ are positive, we can rewrite the inequality as:

$$
\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{\sqrt{2a_i}^2}{\sqrt{2b_i}}
$$

#### Step 3: Consider the identity:

Now, consider the following identity:

$$
\left( \sum_{i=1}^n x_i \right)^2 = \sum_{i=1}^n x_i^2 + 2 \sum_{1 \leq i < j \leq n} x_i x_j
$$

#### Step 4: Apply the identity

Applying this identity to the numerator, we get:

$$
\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} = \frac{\sum_{i=1}^n (2a_i) + 2 \sum_{1 \leq i < j \leq n} \sqrt{2a_i} \sqrt{2a_j}}{\sum_{i=1}^n \sqrt{2b_i}}
$$

#### Step 5: Consider the following:

Since $2a_i$ and $2b_i$ are even integers, $2 \sum_{1 \leq i < j \leq n} \sqrt{2a_i} \sqrt{2a_j}$ is also even.

Now, we have:

$$
\frac{\sum_{i=1}^n (2a_i)}{\sum_{i=1}^n \sqrt{2b_i}} + \frac{2 \sum_{1 \leq i < j \leq n} \sqrt{2a_i} \sqrt{2a_j}}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{(2a_i)}{\sqrt{2b_i}}
$$

#### Step 6: Simplify the expression
$$
\frac{\sum_{i=1}^n (2a_i)}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{(2a_i)}{\sqrt{2b_i}}
$$

Since both sides of the inequality are non-negative, and the numerator of the left-hand side contains the terms of the denominator of the right-hand side, the inequality holds true. Therefore, Sedrakayan's Lemma is proven for $ u_i $ and $ v_i $ being square roots of even integers.
