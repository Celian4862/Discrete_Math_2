# Directed Graphs

```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 21
# Create a directed graph object
G = nx.DiGraph()

# Add nodes to the graph
G.add_nodes_from([1, 2, 3, 4, 5, 6,7])

# Add edges to the graph
G.add_edge(1, 5)
G.add_edge(5, 2)
G.add_edge(2, 6)
G.add_edge(6, 3)
G.add_edge(3, 7)
G.add_edge(7, 4)
G.add_edge(4, 1)


# Draw the graph
pos = nx.circular_layout(G)
nx.draw(G,pos=pos, with_labels=True)

# Show the plot
plt.show()
```

A directed graph's most prominent feature is its edges, which are arrows, showing a direction from one vertex to another. The example above is a seven-vertex star.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a complete graph with 5 nodes
G = nx.complete_graph(5)

# Seed for reproducible layout
seed = 57
pos = nx.spring_layout(G, seed=seed)

# Draw the graph
nx.draw(G, pos=pos, with_labels=True)
plt.show()
```

Above is a complete graph, with each vertex connected exactly once to all other vertices in the graph.

```python
import matplotlib.pyplot as plt
import networkx as nx

network = nx.Graph()
network.add_nodes_from(['A','B','C'], bipartite=0) # Add the node attribute "bipartite"
network.add_nodes_from(['1','2', '3'], bipartite=1)
network.add_edges_from([('A', '1'), ('A', '2'), ('A', '3'),
                              ('B', '1'), ('B', '2'), ('B', '3'),
                              ('C', '1'), ('C', '2'), ('C', '3')])

top = nx.bipartite.sets(network)[0]
pos = nx.bipartite_layout(network, top)
nx.draw(network, pos=pos, with_labels=True, node_color=['pink','magenta','red','deepskyblue','cyan', 'green'])
plt.show()
```

Above is a bipartite graph, with two sets of vertices. Each vertex in one set must connect exactly once to all vertices in the other set, but it cannot connect to a vertex in its own set.
