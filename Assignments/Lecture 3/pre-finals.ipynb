# Bipartite Graph

```python
import matplotlib.pyplot as plt
import networkx as nx

network = nx.Graph()
network.add_nodes_from(['A','B','C'], bipartite=0) # Add the node attribute "bipartite"
network.add_nodes_from(['1','2', '3'], bipartite=1)
network.add_edges_from([('A', '1'), ('A', '2'), ('A', '3'),
                              ('B', '1'), ('B', '2'), ('B', '3'),
                              ('C', '1'), ('C', '2'), ('C', '3')])

top = nx.bipartite.sets(network)[0]
pos = nx.bipartite_layout(network, top)
nx.draw(network, pos=pos, with_labels=True, node_color=['pink','magenta','red','deepskyblue','cyan', 'green'])
plt.show()
```

Above is a bipartite graph, with two sets of vertices. Each vertex in one set must connect exactly once to all vertices in the other set, but it cannot connect to a vertex in its own set.

# Degree of a Vertex

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 0
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from([1,2,3,4,5,6,7])

edgelist = [(1,2),(1,3),(4,5), (2, 4), (1, 1), (4, 6), (1, 4), (7, 7)]
G.add_edges_from(edgelist)
pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

A degree of a vertex is determined by how many edge endpoints it has. Every edge that connects two vertices adds one degree to both. Because a loop has both its endpoints on the same vertex, it adds two degrees instead of one.

# Travelling in a Graph

## Walk

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty undirected graph
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from([1, 2, 3, 4, 5])

# Add edges to the graph
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 1)])

# Draw the graph
pos = nx.spring_layout(G)  # Positioning nodes
nx.draw(G, pos, with_labels=True)
walk = [(5, 1),(1, 2),(2, 3), (1, 3)]

# Draw the graph with edge colors
nx.draw_networkx_edges(G, pos=pos, edgelist = walk,edge_color='cyan',width=2)

# Display the plot
plt.show()
```

Shown above is a walk. A walk is essentially a subgraph with a finite set of ordered vertices and edges, with a defined beginning and end.

## Trail

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty undirected graph
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])

# Add edges to the graph
edgelist = [(1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 1), (6, 1), (6, 5), (7, 5), (7, 4), (8, 4), (8, 3)]
G.add_edges_from(edgelist)

# Define the walk as a list of edges
trail = [(7, 5), (5, 1), (1, 2), (2, 3), (8, 3)]

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos=pos, with_labels=True)

# Draw the edges in the walk with red color
nx.draw_networkx_edges(G, pos=pos, edgelist=trail, edge_color='cyan',width=2)

# Display the plot
plt.show()
```

A trail is like a walk, but it cannot repeat any edges.

# Traveling Salesman Problem

The following is an example of the traveling salesman problem:

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A","B","C","D"])
# Add self-loops to the remaining nodes
edgelist = [("A","B"),("A","C"),("A","D"),("B","C"),("B","D"),("C","D"),("A","C"),("B","D")]
G.add_edges_from(edgelist)
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={('A', 'B'): '30', 
                 ('A', 'C'): '50', 
                 ('A', 'D'): '40',
                 ('B', 'C'): '30',
                 ('B', 'D'): '35',
                 ('C', 'D'): '25',},
    font_color='red'
)
nx.draw(G, pos=pos,with_labels=True)
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    with_labels=True)

plt.show()
```

In a traveling salesman problem, one must find the shortest Hamiltonian circuit, which is based on the weights on the edges (which act as distances).

```python
import matplotlib.pyplot as plt
import networkx as nx

node_names = ["A", "B", "C", "D"]
edges = [('A', 'B', {'weight': 30}),
         ('A', 'C', {'weight': 50}),
         ('A', 'D', {'weight': 40}),
         ('B', 'C', {'weight': 30}),
         ('B', 'D', {'weight': 35}),
         ('C', 'D', {'weight': 25})]

routes = [
    ['A', 'B', 'C', 'D', 'A'],
    ['A', 'B', 'D', 'C', 'A'],
    ['A', 'C', 'B', 'D', 'A'],
    ['A', 'C', 'D', 'B', 'A'],
    ['A', 'D', 'B', 'C', 'A']
]

fig, axs = plt.subplots(1, 5, figsize=(20, 4))

for i, ax in enumerate(axs):
    G = nx.Graph()
    G.add_nodes_from(node_names)
    G.add_edges_from(edges)
    pos = nx.planar_layout(G)

    nx.draw_networkx_edge_labels(
        G, pos,
        edge_labels={edge[:2]: str(edge[2]['weight']) for edge in edges},
        font_color='red', ax=ax
    )

    nx.draw(G, pos=pos, with_labels=True, node_size=700, ax=ax)

    # Highlighting the specified route
    for j in range(len(routes[i]) - 1):
        nx.draw_networkx_edges(
            G, pos, edgelist=[(routes[i][j], routes[i][j + 1])],
            edge_color='cyan', width=2, ax=ax
        )

    # Calculating the weight of the route
    weight = sum(G[routes[i][j]][routes[i][j + 1]]['weight'] for j in range(len(routes[i]) - 1))
    ax.set_title('Route {}'.format(i + 1))
    ax.text(0.5, -0.15, 'Weight: {}'.format(weight), horizontalalignment='center', verticalalignment='center',
            transform=ax.transAxes)

# Adjust layout
plt.tight_layout()
plt.gcf().canvas.manager.set_window_title("TRAVELLING SALESMAN PROBLEM")
plt.show()
```

From the above graphs, one can tell that the shortest path is Route 1.
